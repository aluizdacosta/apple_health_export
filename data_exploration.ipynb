```python
# Iterative approach to find InBody records
inbody_records = []

# Create an iterative parser
context = etree.iterparse('export.xml', events=('end',), tag='Record')

# Process each Record element
for event, elem in context:
    # Get all attributes of this record
    attributes = dict(elem.attrib)
    
    # Check if this is from InBody (checking source or device name)
    is_inbody = False
    for attr_name, attr_value in attributes.items():
        if 'source' in attr_name.lower() or 'device' in attr_name.lower():
            if 'inbody' in str(attr_value).lower():
                is_inbody = True
                break
    
    # If this is an InBody record, save it
    if is_inbody:
        inbody_records.append(attributes)
    
    # Clear element to free memory
    elem.clear()
    while elem.getprevious() is not None:
        del elem.getparent()[0]

# Convert to DataFrame
inbody_df = pd.DataFrame(inbody_records)
print(f"Found {len(inbody_records)} InBody records")
```
